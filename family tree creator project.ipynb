```python
# Importing tkinter
from tkinter import *

# Importing ttk from tkinter
from tkinter import ttk

# Creating app window
app = Tk()

# Defining title of the app
app.title("GUI")

# Defining label of the app and calling a geometry
ttk.Label(app, text ="Treeview(hierarchical)",font = ('Times New Roman', 30)).pack()

# Creating treeview window
treeview = ttk.Treeview(app)

# Calling pack method on the treeview
treeview.pack(fill = BOTH, expand = 1)

# Inserting items to the treeview
# Inserting parent
treeview.insert('', '0', 'item1',text ='FAMILY TREE')

# Inserting child
treeview.insert('', '1', 'item2',text ='Maternal grandparents')
treeview.insert('', '2', 'item3',text =' Paternal grandparents ')


# Inserting more than one attribute of an item

treeview.insert('item2', 'end', 'GrandFather',text ='GrandFather')
treeview.insert('item2', 'end', 'GrandMother',text ='GrandMother')
treeview.insert('item2', 'end', 'Uncle',text ='Uncle')
treeview.insert('item2', 'end', 'Aunt',text ='Aunt')
treeview.insert('item2', 'end', 'father',text ='father')
treeview.insert('item2', 'end', 'mother',text ='mother')
treeview.insert('item2', 'end', 'sister',text ='sister')
treeview.insert('item2', 'end', 'brother',text ='brother')
treeview.insert('item2', 'end', 'cousins',text ='cousins')
treeview.insert('item2', 'end', 'Me',text ='Me')

treeview.insert('item3', 'end', 'Grand Father',text ='Grand Father')
treeview.insert('item3', 'end', 'Grand mother',text ='Grand Mother')
treeview.insert('item3', 'end', 'Elder Father',text ='Elder Father')
treeview.insert('item3', 'end', 'Elder mother',text ='Elder Mother')
treeview.insert('item3', 'end', 'uncle',text ='uncle')
treeview.insert('item3', 'end', 'Aunty',text ='Aunty')
treeview.insert('item3', 'end', 'Father',text ='Father')
treeview.insert('item3', 'end', 'Mother',text ='Mother')
treeview.insert('item3', 'end', 'Sister',text ='Sister')
treeview.insert('item3', 'end', 'Brother',text ='Brother')
treeview.insert('item3', 'end', 'Cousins',text ='Cousins')
treeview.insert('item3', 'end', 'ME',text ='ME')

# Placing each child items in parent widget
treeview.move('item2', 'item1', 'end')
treeview.move('item3', 'item1', 'end')

# Calling main()
app.mainloop()
```
